```python
#謝店長,凡哥,寬哥
import requests
from urllib import request
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import json
import re
import os
import csv
from sklearn import datasets
import openpyxl
#import DataFrame import df
```

```python
user_agent ='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'

headers = {'User-Agent':user_agent}

path=r'./CARDU'
if not os.path.exists(path):
    os.mkdir(path)
```

```python
from itertools import zip_longest
aa=[]
# datafram = pd.DataFrame( columns=['標題','發布時間','網址','內文'])
datafram = pd.DataFrame( columns=['標題','發布時間','網址','內文'])
index = 0
for i in range(1,28):

    url='https://www.cardu.com.tw/card/new_card_list.php?PageNo='+str(i)
    res=requests.get(url,headers=headers)
    soup =BeautifulSoup(res.text,'html.parser')
    article = soup.select('div [class="mb-2"]')
    href_t=soup.select('h3 [href]')
    t=0
    for h in article:

        time=soup.select('small[class="px-md-4 pb-md-0 px-0 pb-2"]')
        time=str(time[t]).replace('<small class="px-md-4 pb-md-0 px-0 pb-2\">',"").replace('</small>',"").replace('(','').replace(')','')
        href=h.a['href']
        res_s=requests.get(href,headers=headers)
        soup_s =BeautifulSoup(res_s.text,'html.parser')
        article_title =soup_s.select('div[id= "news_content"] div[class="pt-3 mx-3 detail_title"] h2')
        article_list = soup_s.select('div[id= "news_content"] div[class = "pb-3 mx-3 detail_content"] p')
        article_href = soup_s.select('meta[property="og:url"]')[0]
        article_href=str(article_href['content'])
#         for i in range(len(article_list)):

        body_yes = ''.join(www.text for www in article_list)

        title = str(article_title).replace('\n','').replace('<h2>','').replace('<h2>','').replace('</h2>','').replace('[','').replace(']','')

        #print(G5_title)
        space= []
        space.append(title)
        space.append(time)
        space.append(article_href)
        space.append(body_yes.replace('\n',''))
        datafram.loc[index] = space
        index+=1
        t+=1
```

```python
datafram.to_excel('CARUD.xls')
```
