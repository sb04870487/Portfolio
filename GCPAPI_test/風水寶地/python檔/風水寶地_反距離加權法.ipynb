```python
# 定義距離算法
from math import radians, cos, sin, asin, sqrt,atan,tan,acos
def getDistance(latA, lonA, latB, lonB):  
        ra = 6378140  # radius of equator: meter  
        rb = 6356755  # radius of polar: meter  
        flatten = (ra - rb) / ra  # Partial rate of the earth  
        # change angle to radians  
        radLatA = radians(latA)  
        radLonA = radians(lonA)  
        radLatB = radians(latB)  
        radLonB = radians(lonB)  
      
        pA = atan(rb / ra * tan(radLatA))  
        pB = atan(rb / ra * tan(radLatB))  
        x = acos(sin(pA) * sin(pB) + cos(pA) * cos(pB) * cos(radLonA - radLonB))  
        c1 = (sin(x) - x) * (sin(pA) + sin(pB))**2 / (cos(x / 2)+0.000000000001)**2  
        c2 = (sin(x) + x) * (sin(pA) - sin(pB))**2 / (sin(x / 2)+0.000000000001)**2  
        dr = flatten / 8 * (c1 - c2)  
        distance = ra * (x + dr)  
        return distance  
```

```python
#sigmoid函數
def sigmoid(x):
    ad = 1. / (1 + np.exp(-x))
    return ad
```

```python
#sigmoid函數_些微調整
def adjust_sigmoid(Inverse_Distance_Sum):
    adjust = (sigmoid(Inverse_Distance_Sum)-0.5)*2
    return adjust
```
